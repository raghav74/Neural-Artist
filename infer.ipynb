```python
from keras.models import load_model

gen=load_model('models/generator.h5')
```

```python
# To make sure that we can reproduce the experiment and get the same results
#np.random.seed(10)

# The dimension of our random noise vector.
random_dim = 100


def plot_generated_images(epoch, generator, examples=25, dim=(5, 5), figsize=(5, 5)):
    noise = np.random.normal(0, 0.5, size=[examples, random_dim])
    #noise=np.zeros((25,100))
    #noise.fill(0.6)
    generated_images = generator.predict(noise)
    generated_images = generated_images.reshape(examples, 28, 28)
    
    %matplotlib inline
    plt.figure(figsize=figsize)
    for i in range(generated_images.shape[0]):
        plt.subplot(dim[0], dim[1], i+1)
        plt.imshow(generated_images[i], interpolation='nearest', cmap='gray_r')
        plt.axis('off')
    plt.tight_layout()
    plt.savefig('imgs/gan_generated_image_epoch_%d.png' % epoch)
```

```python
import numpy as np
import matplotlib.pyplot as plt
plot_generated_images(1, gen)
```
